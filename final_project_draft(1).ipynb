```python
# Load data, define hover text and bubble size
data = pd.read_csv('Interactive Data Final Project.csv')
df_2023 = data['Popular U.S. Cities']

hover_text = []
bubble_size = []
```

```python
for index, row in df_2023.iterrows():
    hover_text.append(('Popular U.S. Cities: {city}<br>'+
                      'Entry-Level Marketing Starting Salary: {salary}<br>'+
                      'Cities Median Income: {income}<br>'+
                      'Cities Population: {pop}<br>'
                      ).format(city=row['city'],
                                            salary=row['salary'],
                                            income=row['income'],
                                            pop=row['pop'],
                                            year=row['year']))
    bubble_size.append(math.sqrt(row['pop']))

```
